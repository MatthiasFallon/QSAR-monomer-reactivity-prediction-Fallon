```python
import pandas as pd
import numpy as np
import seaborn
```

```python
# Import data
full_data = pd.read_table('Vdesc.txt', sep='\t')

# Anything below -0.1984 & Anything above 0.1996
# 84 neg for V, 125 pos for V, 209 total

#Anything below -0.1993 & Anything above 0.1984
#

# Getting U and V values
u_values = full_data['U'].tolist()
v_values = full_data['V'].tolist()
u_cor = []
v_cor = []
list(np.float_(u_values))
list(np.float_(v_values))

des_names = full_data.columns.values.tolist()
des_names = des_names[2:186]
```

```python
print(len(v_cor))
```

```python
# Calculating correlations
for col in des_names:
    des_values = full_data.get(col)
    list(np.float_(des_values))
    u = np.corrcoef(u_values, des_values)
    v = np.corrcoef(v_values, des_values)
    u_cor.append(u[0, 1])
    v_cor.append(v[0, 1])
```

```python
print(len(v_cor))
print(len(des_names))
```

```python
# Splitting up data for manageability of figure sizing

des_cor0 = pd.DataFrame({'V': v_cor[0:19]},
                        index=des_names[0:19])

des_cor1 = pd.DataFrame({'V': v_cor[20:39]},
                        index=des_names[20:39])

des_cor2 = pd.DataFrame({'V': v_cor[40:59]},
                        index=des_names[40:59])

des_cor3 = pd.DataFrame({'V': v_cor[60:79]},
                        index=des_names[60:79])

des_cor4 = pd.DataFrame({'V': v_cor[80:99]},
                        index=des_names[80:99])

des_cor5 = pd.DataFrame({'V': v_cor[100:119]},
                        index=des_names[100:119])

des_cor6 = pd.DataFrame({'V': v_cor[120:139]},
                        index=des_names[120:139])

des_cor7 = pd.DataFrame({'V': v_cor[140:159]},
                        index=des_names[140:159])

des_cor8 = pd.DataFrame({'V': v_cor[160:179]},
                        index=des_names[160:179])

des_cor9 = pd.DataFrame({'V': v_cor[180:184]},
                        index=des_names[180:184])
```

```python
import matplotlib.pyplot as plt
fig, (ax0, ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8, ax9) = plt.subplots(ncols=10, figsize=(30, 6))
fig.subplots_adjust(wspace=1.25)

seaborn.heatmap(des_cor0, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax0)
seaborn.heatmap(des_cor1, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'viridis': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax1)
seaborn.heatmap(des_cor2, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax2)
seaborn.heatmap(des_cor3, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax3)
seaborn.heatmap(des_cor4, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax4)
seaborn.heatmap(des_cor5, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax5)
seaborn.heatmap(des_cor6, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax6)
seaborn.heatmap(des_cor7, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax7)
seaborn.heatmap(des_cor8, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax8)
seaborn.heatmap(des_cor9, annot=True, cmap='viridis',
                cbar=False, cbar_kws={'shrink': 0.5},
                fmt='.2f', annot_kws={'size': 8},
                ax=ax9)
plt.savefig('descriptors.png', dpi=1500)
```
